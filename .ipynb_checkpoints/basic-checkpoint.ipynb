```python
def pca(x, k=2):
    #x is of shape p x n where p are the number of features
    x = x-np.mean(x,axis=0, keepdims=True)
    c = np.cov(x.T)/x.shape[0]
    print('c', c.shape)
    e, u = np.linalg.eig(c)
    print('u', u.shape)
    ii = e.argsort()[::-1]
    e, u = e[ii], u[:,ii]
    xk = x @ u[:,:k]
    print('xk', xk.shape)
    return np.real(xk)
```
